# Structured Output in LangChain

## What is Structured Output?

By default, Large Language Models (LLMs) generate **natural language text**. While this is easy for humans to read, it is often difficult for applications to process programmatically.

**Example (Unstructured Output):**

```text
John is 25 years old and works as a Data Scientist.
```

Suppose we want to store this information in a database or send it to an API. It is much easier if the response is returned in a structured format like:

```json
{
    "name": "John",
    "age": 25,
    "profession": "Data Scientist"
}
```

Structured output ensures that the model returns responses in a predefined format instead of free-form text.

---

# Why is Structured Output Needed?

Structured output is useful when building applications such as:

- Retrieval-Augmented Generation (RAG)
- AI Agents
- API integrations
- Database applications
- Data extraction pipelines
- Information extraction from documents

Instead of manually parsing text, the model directly returns data in the expected format.

---

# Ways to Define Structured Output in LangChain

LangChain supports multiple ways to define the expected output structure.

1. TypedDict
2. Pydantic
3. JSON Schema

All three define the **shape** of the expected response, but they differ in validation capabilities and complexity.

---

# 1. TypedDict

## What is TypedDict?

`TypedDict` is a Python typing construct that specifies the expected keys and value types of a dictionary.

It provides **type hints** but does **not perform runtime validation**.

---

## Example

```python
from typing import TypedDict

class Person(TypedDict):
    name: str
    age: int
    profession: str
```

---

## Using TypedDict in LangChain

```python
structured_model = model.with_structured_output(Person)
```

---

## Expected Output

```python
{
    "name": "John",
    "age": 25,
    "profession": "Data Scientist"
}
```

---

## Advantages

- Very simple
- Lightweight
- Easy to understand
- No additional dependencies

---

## Limitations

- No runtime validation
- Cannot detect incorrect data types
- Best suited for simple applications

---

# 2. Pydantic

## What is Pydantic?

Pydantic is a Python library used for **data validation and parsing**.

Unlike TypedDict, Pydantic validates the generated output before returning it.

It is the **most commonly used approach** for structured output in LangChain.

---

## Example

```python
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    age: int
    profession: str
```

---

## Using Pydantic in LangChain

```python
structured_model = model.with_structured_output(Person)
```

---

## Expected Output

```python
Person(
    name="John",
    age=25,
    profession="Data Scientist"
)
```

---

## Advantages

- Runtime validation
- Automatic type conversion
- Cleaner Python objects
- Better error handling
- Industry standard for production applications

---

## Limitations

- Slightly more verbose than TypedDict
- Additional dependency

---

# 3. JSON Schema

## What is JSON Schema?

JSON Schema is a language-independent specification used to define the structure of JSON objects.

It is widely used in:

- APIs
- OpenAI Function Calling
- Gemini Function Calling
- Tool Calling
- Web Services

---

## Example

```python
schema = {
    "type": "object",
    "properties": {
        "name": {
            "type": "string"
        },
        "age": {
            "type": "integer"
        },
        "profession": {
            "type": "string"
        }
    },
    "required": [
        "name",
        "age",
        "profession"
    ]
}
```

---

## Using JSON Schema in LangChain

```python
structured_model = model.with_structured_output(schema)
```

---

## Expected Output

```json
{
    "name": "John",
    "age": 25,
    "profession": "Data Scientist"
}
```

---

## Advantages

- Language independent
- Works across multiple platforms
- Ideal for API contracts
- Compatible with most LLM providers

---

## Limitations

- More verbose
- Harder to write manually
- Less Pythonic than Pydantic

---

# Comparison

| Feature | TypedDict | Pydantic | JSON Schema |
|----------|-----------|-----------|-------------|
| Purpose | Type Hinting | Validation + Parsing | Universal Schema Definition |
| Runtime Validation | ❌ No | ✅ Yes | ✅ Yes |
| Automatic Type Conversion | ❌ No | ✅ Yes | Depends on implementation |
| Best For | Small projects | Production applications | APIs and Tool Calling |
| Python Specific | ✅ Yes | ✅ Yes | ❌ No |

---

# Which One Should You Use?

### TypedDict

Use when:

- Building simple projects
- Learning LangChain
- Runtime validation is not required

---

### Pydantic

Use when:

- Building production applications
- Creating AI Agents
- Building RAG systems
- Performing data extraction
- Working with APIs

This is the **recommended approach** for most LangChain applications.

---

### JSON Schema

Use when:

- Working with APIs
- Tool Calling
- Function Calling
- Cross-language applications
- Defining strict data contracts

---

# Summary

| Method | Best Use Case |
|----------|--------------|
| TypedDict | Lightweight structured output with type hints |
| Pydantic | Production-ready structured output with validation |
| JSON Schema | Universal schema for APIs, tools, and cross-platform applications |

---

# Key Takeaways

- Structured output makes LLM responses machine-readable.
- It eliminates the need to manually parse natural language.
- LangChain supports multiple schema definitions through `with_structured_output()`.
- **TypedDict** provides simple type hints.
- **Pydantic** adds runtime validation and is the preferred choice for production.
- **JSON Schema** is a language-independent standard commonly used for APIs and tool/function calling.
